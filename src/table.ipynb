```python
def generate_crosstab(df, column, other_column):
    # Tabela de valores absolutos
    table_absolute = pd.crosstab([df[column]], df[other_column],\
                                 rownames=[column],\
                                 colnames=[other_column], margins=True)
    table_absolute.loc[column] = ' '
    table_absolute = table_absolute.reindex(index=[column] + table_absolute.index.tolist()[:-1])

    # Tabela de valores percentuais apenas nas colunas correspondentes
    table_percent = pd.crosstab([df[column]], df[other_column],\
                                rownames=[column],\
                                colnames=[other_column], margins=True, normalize='index') * 100

    # Adicionar sufixo '_PERCENT' para distinguir as colunas percentuais
    table_percent.columns = [f"{col}_PERCENT" for col in table_percent.columns]
    table_percent.loc[column] = ' '
    table_percent = table_percent.reindex(index=[column] + table_percent.index.tolist()[:-1])

    return pd.concat([table_absolute, table_percent], axis=1)

# Lista de colunas que você deseja analisar
colunas_analisadas = ['REGIAO', 'Presença de FEBRE', 'Presença de TOSSE']

# Lista para armazenar as tabelas cruzadas geradas
tabelas = []

# Gerar tabelas cruzadas para cada coluna
for coluna in colunas_analisadas:
    tabela = generate_crosstab(dados, coluna, 'MORTE')
    tabelas.append(tabela)

# Concatenar todas as tabelas
result_table = pd.concat(tabelas)


print(result_table)

```
