```python
#Filtrando linhas apenas para crianças e idosos
def converter_idade(row):
    idade = row['NU_IDADE_N']
    tipo_idade = row['TP_IDADE']
    if tipo_idade == 1:  # Dias
        idade /= 365  # Converter dias para anos
    elif tipo_idade == 2:  # Meses
        idade /= 12  # Converter meses para anos
    return idade

dadosFiltrados = dadosFiltrados.assign(IDADE_ANOS=dadosFiltrados.apply(converter_idade, axis=1))
dadosFiltrados = dadosFiltrados[(dadosFiltrados['IDADE_ANOS'] < 12) | (dadosFiltrados['IDADE_ANOS'] >= 60)]

dadosFiltrados
```

```python
#Filtro "Hospitalizado"
dadosFiltrados = dadosFiltrados.loc[dadosFiltrados['HOSPITAL'] == 1]
dadosFiltrados
```

```python
#Categorizando as faixas etárias
def categorize_age(idade):
    if idade <= 0.5:
        return "0 a 6 meses"
    elif idade <= 1:
        return "6 meses a 1 ano"
    elif idade <= 4:
        return "1 a 4 anos"
    elif idade <= 6:
        return "4 a 6 anos"
    elif idade <= 9:
        return "6 a 9 anos"
    elif idade <= 12:
        return "9 a 12 anos"
    elif idade <= 70:
        return "60 a 70 anos"
    elif idade < 80:
        return "70 a 80 anos"
    else:
        return "80 anos ou mais"

dadosFiltrados['FAIXA_ETARIA'] = dadosFiltrados['IDADE_ANOS'].apply(categorize_age)

dadosFiltrados.head()
```

```python
#Categorizando Regiao
dadosFiltrados['SG_UF_NOT'].unique()

# Dicionário de mapeamento de siglas de estados para regiões
regioes = {
    'MG': 'Sudeste',
    'RJ': 'Sudeste',
    'SP': 'Sudeste',
    'ES': 'Sudeste',
    'AM': 'Norte',
    'TO': 'Norte',
    'MS': 'Centro-Oeste',
    'PE': 'Nordeste',
    'PR': 'Sul',
    'DF': 'Centro-Oeste',
    'CE': 'Nordeste',
    'RS': 'Sul',
    'SC': 'Sul',
    'GO': 'Centro-Oeste',
    'BA': 'Nordeste',
    'PA': 'Norte',
    'SE': 'Nordeste',
    'AP': 'Norte',
    'PI': 'Nordeste',
    'RN': 'Nordeste',
    'MA': 'Nordeste',
    'RO': 'Norte',
    'RR': 'Norte',
    'AL': 'Nordeste',
    'MT': 'Centro-Oeste',
    'PB': 'Nordeste',
    'AC': 'Norte',
    # Adicione mais mapeamentos para outros estados e regiões conforme necessário
}

# Função para mapear as siglas de UF para regiões
def categorizar_regiao(uf):
    if uf in regioes:
        return regioes[uf]
    else:
        return 'Outro'  # Defina a categoria padrão para estados não mapeados

# Aplica a função para criar a nova coluna "Regiao"
dadosFiltrados['REGIAO'] = dadosFiltrados['SG_UF_NOT'].apply(categorizar_regiao)
dadosFiltrados['REGIAO'].unique()
```

```python
# Lista de nomes das colunas de sintomas
sintomas_cols = ['FEBRE', 'TOSSE', 'GARGANTA', 'DISPNEIA', 'DESC_RESP', 'SATURACAO', 'DIARREIA', 'VOMITO', 'DOR_ABD'
                   , 'FADIGA', 'PERD_OLFT', 'PERD_PALA', 'OUTRO_SIN']

# Função para mapear valores
def map_values(value):
    if value == 1:
        return 'Yes'
    elif value == 0:
        return 'No'
    else:
        return 'Missing'

# Aplicar a função a cada coluna de sintomas
for col in sintomas_cols:
    new_col_name = 'Presença de ' + col   # Nome da nova coluna
    dadosPreMiss[new_col_name] = dadosPreMiss[col].apply(lambda x: map_values(x))

linha10 = dadosPreMiss.loc[10]
print(linha10)
```
